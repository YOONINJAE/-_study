```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import seaborn as sns
#from sklearn.datasets import load_iris
```

```python
iris = sns.load_dataset('iris')
#모수통계량 - 데이터가 정규분포를 따를때
iris.groupby('species').aggregate(['mean','std']) #보고자하는 것에대해서 그룹바이
```

```python
iris.groupby('species').aggregate(['median','min','max'])
#비 모수에 대한 통계량 - 데이터가 정규분포를 따르지 않을때
#논문을 보면 두 그래프 다 나와있음 
```

```python
### Result_2. Correlation matrix
iris.corr(method = 'pearson')
M = iris.corr(method = 'pearson')
type(M)

plt.matshow(M, vmin = -1, vmax= 1)
plt.colorbar(shrink = 0.8, aspect =10)

plt.show()
M
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.groupby('cut')['price'].aggregate(['mean','std','sem'])\
diamonds
```

```python
diamonds_2 = diamonds.sample(n=5000)
```

```python
groups = diamonds_2.groupby('color')
print(groups.size())

for name, data in groups: #groupby로 나눈 데이터
  print(name)

fig,ax = plt.subplots(figsize = (8,6))

for name, group in groups:
  ax.plot(group.carat,group.price, label = name
          , marker = 'o',
          linestyle = ''
          ,markersize = 8,
          alpha = 0.6)

plt.legend(loc = 'lower right')
# x1 = diamonds_2['carat']
# y1 = diamonds_2['price']
# plt.scatter(x1,y1,marker = 'o', s=50)
plt.title("Carat vs Price", fontsize=20)
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
#TL씨본
sns.regplot(x='carat',
            y='price',
            hue = 'color',
            s= 100,
            alpha = 0.6,
            data = diamonds_2)

plt.show()
```
