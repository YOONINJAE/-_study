```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(0, 2, 0.2)
plt.plot(x, x, 'bo', label = 'OECD average')
plt.plot(x, x**2, 'rD-', markersize =4 , linewidth=1, label = 'korea')

plt.xlabel('Year')
plt.ylabel('job Growth rate')
plt.xticks([0,1,2], labels = ["2019", "2020","2021"])
plt.yticks([0,1,2,3])

plt.legend()

plt.show()
```

```python
x= np.arange(0, 2, 0.2)
plt.plot(x, x, 'bo', label = 'OECD average')
plt.plot(x, x**2, 'rD-', markersize =4 , linewidth=1, label = 'korea')

plt.xlabel('Year')
plt.ylabel('job Growth rate')
plt.xticks([0,1,2], labels = ["2019", "2020","2021"])
plt.yticks([0,1,2,3])
plt.legend()

plt.tick_params(axis = "x",direction ='in', length = 6, pad = 10,
                labelsize = 8, labelcolor = 'k')

plt.tick_params(axis = 'y',direction = 'inout',length = 10, pad = 6,
                labelsize = 12, labelcolor = 'b', color = 'r')

plt.show()
```

```python
#  Title 설정

x = np.arange(0,2, 0.2)

plt.plot(x,x,'bo')
plt.plot(x,x**2, color = 'k', marker = '*', linewidth = 2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.xlabel('Year')
plt.ylabel('job Growth rate')

my_font = {
    'family' : 'serif',
    'color': 'b',
    'weight' : 'bold',
    'size' : 14
}


title = plt.title('Annual economic state', loc='center', pad = 10, fontdict = my_font)

plt.show()

print(title.get_position())
print(title.get_text()) # 모듈형 z클래스 사용할때
```

```python
## Bar plot

x = np.arange(3)
y = np.array([200,300,400])
mycolor = ['r','g','b']
plt.bar(x, y, width = 0.8, color = mycolor, alpha = 0.4,
        edgecolor = 'w', linewidth = 5,
        yerr = y*0.2, ecolor = 'b', capsize = 5)
plt.xticks(x, labels = ['2019','2020','2021'])
plt.show()
```

```python
## 수평 막대 그래프

y = np.arange(3) # 막대그래프를 그리기 위한 숫자임
values = [100,200,300]
years = ['2019','2020', '2021']
colors = ['y', 'dodgerblue', 'C2'] #싸이클리 컬러 C0 ~ C10

plt.barh(y, values, color = colors, alpha = 0.5
         ,tick_label = years, edgecolor = 'r', linewidth = 3)
plt.show()
```

```python
## 히스토그램 분포에 관심이 있는것 정보는 의미 x 

np.random.seed(910) #seed를 일정하게줘야 같은 값 유지

weight = np.random.normal(loc = 65, size = 50, scale =10)
#weight.max()
print(weight)

n ,bins, patches = plt.hist(weight, bins = 15,edgecolor = 'r', alpha =0.5)
plt.show()
print(n ,bins, patches)
# 데이터 가지고 놀때 n, binsm patches 리턴 값 가져올 수 있다.

 
```

```python
a = np.random.randn(1000)
a1 = 2*np.random.randn(1000) + 1 # 표준편차 전체 데이터에 +1
a2 = 20*np.random.rand(5000)  - 10 # * 폭늘리기

plt.hist(a, bins = 100, density = True, alpha=1, histtype = 'step')
plt.hist(a1, bins = 100, density = True, alpha=0.2, histtype = 'stepfilled')
plt.hist(a2, bins = 100, density = True, alpha=0.2, histtype = 'step')
plt.show()
```

```python
## error bar plot 
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
my_err =  yerr = [2.3, 3.1, 1.7, 2.5]

plt.errorbar(x,y, yerr = my_err, capsize = 4, fmt = '--' )

plt.show()
```

```python
# 파이 차트

ratio = [34, 32, 16, 18]
mylabels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels = mylabels, 
        autopct = '%.1f%%',#소수점 2째자리는 약속~
        
        counterclock = True)

plt.show()
```

```python
ratio = [34, 32, 16, 18]
mylabels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.1, 0, 0] #간지나게 떨어트려 버리기

mcolors = ['silver', 'gold', 'lightfray', 'darkgray']
wedgeprops = {'width': 0.6}

plt.pie(ratio, labels = mylabels, 
        autopct = '%.1f%%',
        explode = explode,#소수점 2째자리는 약속~
        startangle = 90,
        shadow = True,
        pctdistance = 0.7, # 글씨 떨어트려버리기
        #colors = mcolors,
        wedgeprops = wedgeprops,
        counterclock = True)

plt.show()
```

```python
# heatmap

arr = np.random.normal(size = (40,50))
plt.matshow(arr, cmap = 'BuGn') # cmap으로 color 가져와서 사용가능
plt.colorbar(shrink = 0.8, #컬러바 줄이기 확률값 3이상은 rear하다
             aspect = 14, # 폭 늘리기
             )
plt.clim(-2, 2) # color limit를 -1~1로 조정하기
plt.show()
```

```python
np.random.seed(1234)

data_a = np.random.normal(0, 2, 1000)
data_b = np.random.normal(-3, 1.3, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

plt.boxplot([data_a, data_b, data_c],
            whis = 1.5, # 2.5 데이터 전체를 의미있게 (좌우 boundary 상승)사용하겠다 1.5는 뺄수있는 근거.
            notch = True, # 데이터 median value 강조
            vert = False #옆으로 눞힐 수 있음
            )
plt.show()
```
