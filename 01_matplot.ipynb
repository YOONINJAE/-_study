```python
plt.plot(np.arange(1,5), np.arange(1,5)**2)
plt.xlabel('X-label', labelpad = 15,
           fontdict = {'family' : 'serif', 'color': 'r'})
plt.ylabel('Y-label', labelpad = 20,
           fontdict = {'family' : 'fantasy', 'color' : 'b'})
plt.show()
```

```python
font1 = {
    'family' : 'serif',
    'color' : 'b',
    'weight' : 'bold',
    'size' : 12
}
font2 = {
    'family' : 'fantasy',
    'color' : 'pink',
    'weight' : 'normal',
    'size' : 20
}

plt.plot(np.arange(1,5), np.arange(1,5)**2)

plt.xlabel('X-label', fontdict = font1, labelpad = 12)
plt.ylabel('Y-label', fontdict = font2, labelpad = 12)
plt.show()
```

```python
plt.plot(np.arange(5), np.arange(5)**2, label = 'Price') #label이 legend로 푯현 위치파악??

plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.legend(loc= (.1,.8)) # 범례 위치 조정 가능~!~

# 'upper left', 'upper right', 'lower left', 'lower right'
# 'upper center', 'lower center', 'center left', 'center right'
plt.legend(loc = 'lower right')
plt.grid()
plt.show()
```

```python
plt.plot(np.arange (1,5), np.arange(1,5)*2, label = 'Price')
plt.plot(np.arange (1,5), np.arange(1,5)**2, label = 'Demand')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.legend(loc = 'best', ncol = 2, fontsize=10,
           frameon = True, shadow = False,
           facecolor = 'pink', edgecolor ='b', framealpha = 0.2)
plt.show()
```

```python
# 구간 filing  ===== fill_between()  - 두 수평 방향의 곡선 사이
x = [1,2,3,4]
y = [2, 3, 5, 10]

plt.plot(x,y)
plt.xlabel('X-축')
plt.ylabel('Y-축')
#plt.fill_between([1,3],[3,5], alpha = 0.5)
#plt.fill_between(x[1:3],y[1:3], alpha = 0.5)

plt.fill_betweenx(y[1:3], x[1:3], alpha = .5)

plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.xlabel('x-축')
plt.ylabel('y-축')

plt.plot(x,y1, label='zz')
plt.plot(x,y2, label = 'xx')
plt.fill_between([2,3],[3,5],[2,4], alpha = 0.5, color = 'lightgray')
plt.legend(loc = 'best')

plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.fill([2, 2, 3, 3], [1, 3, 6, 4], alpha =0.5, color = 'lightgray')
```
